count: The number of non-empty datapoints that are in that column.

mean: The average value in that column.

std: The standard deviation. The expected amount of variance a datapoint has about the mean. Certain percentages of data within a dataset fall within the 1st (~70%), 2nd (~95%) and 3rd (~99.9%) standard deviation.

min: The lowest value in the column.

25%: The first quartile, that number is the 25th percentile, the lowest 25% of values are below this value. Used in tandem with third quartile to help calculate the interquartile range (IQR) which is used for many things like outliers

50%: The median, the "middle" number or 50th percentile, half of the values in the column are lower than this number. Can be used as an alternate "average" over mean

75%: The third quartile. 75% of values in the column are lower than this number. Used in tandem with first quartile to help calculate the interquartile range (IQR) which is used for many things like outliers.

max: The largest value in the column.

     - **`df.describe()`** provides summary statistics for each column independently, and the count reflects the number of non-missing values in each column across the entire dataset.
     - **`df.groupby("col1")["col2"].describe()`** computes the count for `col2` within each group defined by `col1`, excluding missing values in `col2` for each group.
   - The fundamental difference lies in the context of analysis: `df.describe()` is global, while `df.groupby().describe()` is group-specific.
